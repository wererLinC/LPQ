```python
from sklearn.ensemble import IsolationForest
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

train_x = []
train_y = []
# 读取训练数据集
for i in range(5):
    file = "./dataSet/data_batch_%d"%(i+1)
    dicts = unpickle(file)
    labels = dicts[b'labels']
    data = dicts[b'data'].reshape(-1, 3, 32, 32) # 3*32x32
    data = data.transpose(0,2,3,1)

    for j in range(len(labels)):
        img_data = data[j]
        label = labels[j]
        train_x.append(img_data)
        train_y.append(label)
# 读取测试数据集
test_x = []
test_y = []
for i in range(1):
    # file = "./data_batch_%d"%(i+1)
    file = "./dataSet/test_batch"
    dicts = unpickle(file)
    labels = dicts[b'labels']
    data = dicts[b'data'].reshape(-1, 3, 32, 32) # 3*32x32
    data = data.transpose(0,2,3,1)

    for j in range(len(labels)):
        img_data = data[j]
        label = labels[j]
        test_x.append(img_data)
        test_y.append(label)
train_x = np.array(train_x)
train_y = np.array(train_y)
test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
import time
from sklearn.metrics import roc_curve, auc, roc_auc_score    

for nomal_label in range(10):
    print("################# 正常样本是 %d ################"%(nomal_label))
    norm_train_x = train_x[train_y == nomal_label]
    norm_train_x = norm_train_x.flatten()
    norm_train_x = norm_train_x.reshape(-1, 3072)
    
    test_x = test_x.flatten()
    test_x = test_x.reshape(-1, 3072)

    clf = IsolationForest(n_estimators=100, max_samples='auto').fit(norm_train_x)
    start_time = time.time()
    y_predict = clf.predict(test_x)
    print("消耗的时间为:",(time.time()-start_time)/test_x.shape[0]*1000)
    
    true_y = test_y.copy()
    true_y[test_y == nomal_label] = 1
    true_y[test_y != nomal_label] = 0

    y_predict = y_predict.reshape((-1, 1))
    true_y = true_y.reshape((-1, 1))
    
    score = clf.decision_function(test_x)
    fpr, tpr, tresholds = roc_curve(true_y.flatten(), score.flatten())
    print("AUROC is :", auc(fpr, tpr))
```
